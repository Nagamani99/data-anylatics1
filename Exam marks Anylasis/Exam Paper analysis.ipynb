# INTRODUCTION

 The dataset titled "class_marks.csv" contains student performance data for an academic assessment. It records marks obtained by students in various questions that together contribute to their total score.

```python
import pandas as pd
a=pd.read_csv("class_marks.csv")
import matplotlib.pyplot as plt
an
```

```python
a.shape
```

```python
a["Q1"] = a["Q1aM4"] + a["Q1bM6"]
a["Q2"] = a["Q2aM6"] + a["Q2bM4"]
a["Q3"] = a["Q3aM5"] + a["Q3bM5"]
a["Q4"] = a["Q4aM3"] + a["Q4bM7"]
a["Q5"] = a["Q5M10"]
a["Q6"] = a["Q6aM4"] + a["Q6bM6"]
a.drop(["Q1aM4", "Q1bM6","Q2aM6","Q2bM4" ,"Q3aM5" , "Q3bM5", "Q4aM3" ,"Q4bM7","Q5M10","Q6aM4","Q6bM6"], axis=1, inplace=True) 
a
```

```python
a["Percentage"]=a.Total/60 *100
a.fillna(0,inplace=True)
```

```python
a
```

```python
def assign_grade(percentage):
    if percentage >= 61:
        return 'O'
    elif percentage >= 55:
        return 'A+'
    elif percentage >= 45:
        return 'A'
    elif percentage >= 30:
        return 'B'
    else:
        return 'F'

a['Grade'] = a['Percentage'].apply(assign_grade)

a
```

```python
a
```

```python
grade_counts = a["Grade"].value_counts()
plt.figure(figsize=(8, 5))
grade_counts.plot(kind="bar", color="skyblue")
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.xticks(rotation=0)
plt.show()
```
