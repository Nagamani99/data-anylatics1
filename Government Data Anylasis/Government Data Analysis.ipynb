# Health costs CPI comparison between sectors.Insight: Urban areas might have slightly higher average indices due to healthcare service availability or pricing.

```python
from scipy.stats import ttest_ind
rural_health = df[df['Sector'] == 'Rural']['Health']
urban_health = df[df['Sector'] == 'Urban']['Health']

t_stat, p_value = ttest_ind(rural_health, urban_health)

print("T-test Results for Health CPI (Rural vs Urban):")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Significant difference between Rural and Urban Health CPI.")
else:
    print("Conclusion: No significant difference between Rural and Urban Health CPI.")
```

# t-test for 'Health' CPI between Rural and Urban

```python
from scipy.stats import ttest_ind
rural_health = df[df['Sector'] == 'Rural']['Health']
urban_health = df[df['Sector'] == 'Urban']['Health']

t_stat_ind, p_value_ind = ttest_ind(rural_health, urban_health)

print("Independent t-test (Rural vs Urban on Health CPI):")
print(f"T-statistic: {t_stat_ind:.4f}")
print(f"P-value: {p_value_ind:.4f}\n")
```

# Independent t-test (Rural vs Urban on CPI Health component)

```python
from scipy.stats import ttest_rel
ru_data = df[df['Sector'] == 'Rural+Urban']
t_stat_rel, p_value_rel = ttest_rel(ru_data['Health'], ru_data['Education'])
print("Paired t-test (Health vs Education CPI in Rural+Urban sector):")
print(f"T-statistic: {t_stat_rel:.4f}")
print(f"P-value: {p_value_rel:.4f}\n")
```

# Paired t-test (Rural+Urban sector, comparing Health vs Education CPI)

```python
from scipy.stats import ttest_1samp

urban_health = df[df['Sector'] == 'Urban']['Health']

t_stat_one, p_value_one = ttest_1samp(urban_health, popmean=140)

print("One-sample t-test (Urban Health CPI vs population mean = 140):")
print(f"T-statistic: {t_stat_one:.4f}")
print(f"P-value: {p_value_one:.4f}\n")
```

```python
from scipy.stats import chi2_contingency

bins = [0, 125, 140, 160]
labels = ['Low', 'Medium', 'High']

df['Health_Bucket'] = pd.cut(df['Health'], bins=bins, labels=labels)
contingency = pd.crosstab(df['Sector'], df['Health_Bucket'])

print("\nContingency Table:")
print(contingency)

chi2_stat, chi2_p, dof, expected = chi2_contingency(contingency)

print("\nChi-Square Test Results:")
print(f"Chi2 Stat: {chi2_stat:.4f}")
print(f"P-value: {chi2_p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(expected)

if chi2_p < 0.05:
    print("Conclusion: Significant association between Sector and Health CPI Bucket.")
else:
    print("Conclusion: No significant association between Sector and Health CPI Bucket.")
```
